```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading my csv file
df=pd.read_csv('/content/INDIA_685.csv')
```

```python
# printing first five row
df.head()
```

```python
#checking the last five row
df.tail()
```

```python
#checking for missing value

df.isnull().sum()
```

```python
# Fill missing values in the 'Patient status' column with the mode
mode_value = df['Patient status'].mode()[0]
df['Patient status'].fillna(mode_value, inplace=True)
```

**BAR Graph**

To create a bar graph showing the distribution of patient genders in your dataset:

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Patient Genders')
plt.show()
```

**Heatmap**

To create a heatmap of the correlations between numeric variables in your dataset:

```python

numeric_cols = ['Patient age']

# Remove rows with missing (NaN) values in the 'Patient age' column
df = df.dropna(subset=numeric_cols)

# Check if the 'Patient age' column is numeric, and convert it if necessary
df['Patient age'] = pd.to_numeric(df['Patient age'], errors='coerce')

# Calculate the correlation matrix
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Vilon plot**

To create a violin plot of the distribution of patient ages by gender:

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Gender', y='Patient age', palette='Set3')
plt.title('Distribution of Patient Ages by Gender')
plt.show()
```

**Pie Char**
To create a pie chart showing the distribution of patient statuses in your dataset:

```python
status_counts = df['Patient status'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Patient Statuses')
plt.show()
```

**Scatter Plot:**



```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Collection_date'], df['Patient age'], c='blue', alpha=0.5)
plt.xlabel('Collection Date')
plt.ylabel('Patient Age')
plt.title('Scatter Plot of Patient Age vs. Collection Date')
plt.grid(True)
plt.show()
```

**Box plot**


To visualize the distribution of patient ages by gender using a box plot:

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Gender', y='Patient age', palette='Set2')
plt.title('Distribution of Patient Ages by Gender (Box Plot)')
plt.show()
```

**Histogram:**

To visualize the distribution of patient ages using a histogram:

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Patient age'], bins=20, color='skyblue', alpha=0.7)
plt.xlabel('Patient Age')
plt.ylabel('Frequency')
plt.title('Distribution of Patient Ages (Histogram)')
plt.show()
```
