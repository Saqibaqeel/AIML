```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,3)
plt.show()
```

```python
plt.quiver(0,0,4,3,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
```

ploting multiple vector in single plot


```python
plt.quiver(0,0,4,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,-6,-3,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

addition of vector

```python
vectoe_1=np.array([0,0,2,3])
vectoe_2=np.array([0,0,2,-3])
sum=vectoe_1+vectoe_2
print(sum)
```

plotint the sum

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,2,-3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,0,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
```
