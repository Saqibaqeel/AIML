```python
#importing matplotlib
import matplotlib.pyplot as plt
#importing matplotlib
# import matplotlib as plt
```

```python
#import numpy
import numpy as np
```

```python
x=np.linspace(0,1,100)
y=np.sin(x)
z=np.cos(x)

print(x)
print(y)
print(z)
```

ploting data


```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x ,z)
plt.show()
```

```python
#adding titlex- axis and y -axis label
plt.plot(x,y)
plt.ylabel("sine  value")
plt.xlabel("angle value")
plt.title("sine wave")
plt.show()
```

```python
#parabola
x=np.linspace(-9,10,30)
```

```python
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')#g represent green color and + represent line
plt.show()
```

```python
plt.plot(x,y,'r*')#g represent green color and + represent line
plt.show()
```

ploting multiple graph in single plot



```python
x=np.linspace(-5,10,50)
plt.plot(x,np.cos(x),'g+')
plt.plot(x,np.sin(x),'r--')
plt.show()
```

BAR plot

```python
fig=plt.figure()
```

```python
fig=plt.figure()

axis=fig.add_axes([0,0 ,1,1])
languages=['eng','hind','urdu','bengali','assamiya']
people=[20,40,70,60,10]
axis.bar(languages,people)
plt.xlabel("languages")
plt.ylabel("people")
plt.show()
```

PIE Chart

```python
fig=plt.figure()

axis=fig.add_axes([0,0 ,1,1])
languages=['eng','hind','urdu','bengali','assamiya']
people=[20,40,70,60,10]
axis.pie(people,labels=languages ,autopct='%1.1f%%')
plt.show()
```

Scatter plot


```python
x=np.linspace(5,1,50)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.scatter(x,y,color='r')
axis.scatter(x,z,color='g')
plt.show()
```

3d scatter plot

```python
fig=plt.figure()
axis=plt.axes(projection='3d')
x=20*np.random.random(200)
y=np.sin(x)
z=np.cos(x)
axis.scatter(x,y,z,cmap='green')
plt.show()
```
